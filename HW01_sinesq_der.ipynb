This notebook is going to plot the $\sin^2{\frac{1}{x(2-x)}}$, its derivative, and its second derivative.

```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
import myder
```

```python
x1 = np.arange(-4, -0.01,0.01)
x2 = np.arange(0.01,1.99,0.01)
x3 = np.arange(2.01,6,0.01)
x = np.append(x1,x2)
x = np.append(x,x3)

nx = len(x)
y = lambda x: np.sin(1 / (x * (2 - x)))**2
funy = np.vectorize(y)
print(type(y))
print(type(funy))
```

## Plot function $f(x) = \sin^2{\frac{1}{x(2-x)}}$ in range (0,2) and (2,6) 

```python
plt.figure()
plt.plot(x,funy(x))
plt.xlabel('x')
plt.ylabel('f(x)')
```

## Calculate its first derivative using myder.py

```python
# Plot first derivative using centraldiff

cdy = []
cdy = array(myder.centraldiff(y,-4,6,nx))
```

```python
plt.figure()
plt.plot(x,cdy)
plt.xlabel('x')
plt.ylabel('df/dx')

plt.figure()
plt.plot(x,cdy)
plt.xlim(-0.3,2.3)
plt.xlabel('x')
plt.ylabel('df/dx')

```

```python
# Compare different methods of calculating first derivative
fdy = []
bdy = []
cdy = []
hody = []

fdy = array(myder.forwardiff(y,-4,6,nx+1))
bdy = array(myder.backwardiff(y,-4,6,nx+1))
cdy = array(myder.centraldiff(y,-4,6,nx))
hody = array(myder.centraldiff(y,-4,6,nx))

fdy_cdy = fdy - cdy
bdy_cdy = bdy - cdy
hody_cdy = hody - cdy
```

```python
plt.figure()
plt.plot(x,fdy_cdy)
plt.xlim(-0.3,2.3)
plt.xlabel('x')
plt.ylabel('fdy_cdy')

plt.figure()
plt.plot(x,bdy_cdy)
plt.xlim(-0.3,2.3)
plt.xlabel('x')
plt.ylabel('bdy_cdy')

plt.figure()
plt.plot(x,hody_cdy)
plt.xlim(-0.3,2.3)
plt.xlabel('x')
plt.ylabel('hody_cdy')
```

## Calculate its second derivative

```python
cddy = array(myder.central2diff(y,-4,6,nx))
```

```python
plt.figure()
plt.plot(x,cddy)
plt.xlim(-0.3,2.3)
plt.xlabel('x')
plt.ylabel('d2f/dx2')
```
